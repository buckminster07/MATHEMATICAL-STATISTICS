```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import fitter

# Load the data
file_path = "/content/Copy of UNIT 1 H&F - Final Attendance 2025 April.xlsx"
sheet1 = pd.read_excel(file_path, sheet_name="Sheet1", header=2)  # Skip first 2 rows
sheet2 = pd.read_excel(file_path, sheet_name="Sheet2")

# Clean Sheet1 data
# Remove rows with all NaN (empty rows)
sheet1 = sheet1.dropna(how='all')
# Remove columns with all NaN (empty columns)
sheet1 = sheet1.dropna(axis=1, how='all')

# Convert "Yes"/"No" to 1/0 for attendance
attendance_cols = sheet1.columns[3:-4]  # Columns D to Y (assuming structure)
for col in attendance_cols:
    sheet1[col] = sheet1[col].map({'Yes': 1, 'No': 0})

# Calculate attendance percentage for each student if not already present
if 'Percentage %' not in sheet1.columns:
    sheet1['Percentage %'] = sheet1[attendance_cols].mean(axis=1) * 100

# Prepare data for distribution fitting
attendance_percentages = sheet1['Percentage %']
total_presents = sheet1[attendance_cols].sum(axis=1)
total_sessions = len(attendance_cols)

```

```python
# Set style for plots
#plt.style.use('seaborn')
plt.figure(figsize=(15, 10))

# Histogram of attendance percentages
plt.subplot(2, 2, 1)
sns.histplot(attendance_percentages, kde=True, bins=15)
plt.title('Distribution of Attendance Percentages')
plt.xlabel('Attendance Percentage')
plt.ylabel('Frequency')

# Boxplot of attendance percentages
plt.subplot(2, 2, 2)
sns.boxplot(x=attendance_percentages)
plt.title('Boxplot of Attendance Percentages')
plt.xlabel('Attendance Percentage')

# QQ plot to check normality
plt.subplot(2, 2, 3)
stats.probplot(attendance_percentages, dist="norm", plot=plt)
plt.title('Q-Q Plot (Normal Distribution)')

# ECDF plot
plt.subplot(2, 2, 4)
sns.ecdfplot(attendance_percentages)
plt.title('Empirical Cumulative Distribution Function')
plt.xlabel('Attendance Percentage')
plt.ylabel('Probability')

plt.tight_layout()
plt.show()

# Basic statistics
print("Descriptive Statistics:")
print(attendance_percentages.describe())
print("\nSkewness:", attendance_percentages.skew())
print("Kurtosis:", attendance_percentages.kurtosis())
```

```python

import pandas as pd
import numpy as np
from scipy.stats import beta, gamma, lognorm, norm, kstest
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.diagnostic import lilliefors

# Load Excel file
df = pd.read_excel("Copy of UNIT 1 H&F - Final Attendance 2025 April.xlsx", sheet_name="Sheet2")
df = df.iloc[:, 1:]  # Remove index/label column if needed

# Flatten all roll numbers and remove NaNs
all_rolls = df.values.flatten()
all_rolls = all_rolls[~pd.isnull(all_rolls)]

# Calculate attendance
total_sessions = df.shape[1]
attendance_counts = pd.Series(all_rolls).value_counts().sort_index()
attendance_percentages = (attendance_counts / total_sessions) * 100

# Normalize for beta fit (strictly between 0 and 1)
normalized = (attendance_percentages - attendance_percentages.min()) / (attendance_percentages.max() - attendance_percentages.min())
normalized = normalized.clip(1e-4, 1 - 1e-4)

# Fit distributions
params_norm = norm.fit(attendance_percentages)
params_beta = beta.fit(normalized, floc=0, fscale=1)
params_gamma = gamma.fit(attendance_percentages)
params_lognorm = lognorm.fit(attendance_percentages, floc=0)

# Run KS tests
ks_norm = kstest(attendance_percentages, 'norm', args=params_norm)
ks_beta = kstest(normalized, 'beta', args=params_beta)
ks_gamma = kstest(attendance_percentages, 'gamma', args=params_gamma)
ks_lognorm = kstest(attendance_percentages, 'lognorm', args=params_lognorm)

# Display results
print("\nKolmogorov–Smirnov Test Results:")
print(f"Normal:    KS Statistic = {ks_norm.statistic:.4f}, p-value = {ks_norm.pvalue:.4f}")
print(f"Beta:      KS Statistic = {ks_beta.statistic:.4f}, p-value = {ks_beta.pvalue:.4f}")
print(f"Gamma:     KS Statistic = {ks_gamma.statistic:.4f}, p-value = {ks_gamma.pvalue:.4f}")
print(f"Lognormal: KS Statistic = {ks_lognorm.statistic:.4f}, p-value = {ks_lognorm.pvalue:.4f}")

# Lilliefors test (for normality)
lss_stat, lss_p = lilliefors(attendance_percentages)
print(f"\nLilliefors Test for Normality: Statistic = {lss_stat:.4f}, p-value = {lss_p:.4f}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from fitter import Fitter
import scipy.stats as stats

# Load your data (replace with actual data)
attendance_percentages = pd.Series([0.75, 0.80, 0.85, 0.90, 0.95, 0.70, 0.65, 0.88, 0.92, 0.78])

# Prepare data (remove NaNs)
data = attendance_percentages.dropna().values

# Initialize Fitter
f = Fitter(data, distributions=['gamma', 'lognorm', 'beta', 'norm', 'expon', 'weibull_min', 'weibull_max', 'uniform'])
f.fit()

# Get best fits (lowest SSE)
best_fits = f.get_best(method='sumsquare_error')
best_dist_name = list(best_fits.keys())[0]
best_params = best_fits[best_dist_name]

print("Best-fitting distribution:", best_dist_name)
print("Parameters:", best_params)

# Perform KS, AD, and Chi-Square tests
def goodness_of_fit_tests(data, dist_name, params):
    # Kolmogorov-Smirnov Test
    if dist_name == 'gamma':
        ks_stat, ks_p = stats.kstest(data, 'gamma', args=(params['a'], params['loc'], params['scale']))
    elif dist_name == 'lognorm':
        ks_stat, ks_p = stats.kstest(data, 'lognorm', args=(params['s'], params['loc'], params['scale']))
    elif dist_name == 'beta':
        ks_stat, ks_p = stats.kstest(data, 'beta', args=(params['a'], params['b'], params['loc'], params['scale']))
    elif dist_name == 'norm':
        ks_stat, ks_p = stats.kstest(data, 'norm', args=(params['loc'], params['scale']))
    elif dist_name == 'expon':
        ks_stat, ks_p = stats.kstest(data, 'expon', args=(params['loc'], params['scale']))
    elif dist_name == 'weibull_min':
        ks_stat, ks_p = stats.kstest(data, 'weibull_min', args=(params['c'], params['loc'], params['scale']))
    elif dist_name == 'weibull_max':
        ks_stat, ks_p = stats.kstest(data, 'weibull_max', args=(params['c'], params['loc'], params['scale']))
    elif dist_name == 'uniform':
        ks_stat, ks_p = stats.kstest(data, 'uniform', args=(params['loc'], params['scale']))

    # Anderson-Darling Test (for normality, but adapted for other distributions)
    if dist_name == 'norm':
        ad_stat, ad_crit_vals, ad_sig_levels = stats.anderson(data, dist='norm')
    else:
        ad_stat, ad_crit_vals, ad_sig_levels = None, None, None

    # Chi-Square Test (requires binning)
    hist, bin_edges = np.histogram(data, bins='auto', density=True)
    cdf = np.cumsum(hist * np.diff(bin_edges))
    if dist_name == 'gamma':
        expected = stats.gamma.cdf(bin_edges[1:], a=params['a'], loc=params['loc'], scale=params['scale'])
    elif dist_name == 'beta':
        expected = stats.beta.cdf(bin_edges[1:], a=params['a'], b=params['b'], loc=params['loc'], scale=params['scale'])
    elif dist_name == 'norm':
        expected = stats.norm.cdf(bin_edges[1:], loc=params['loc'], scale=params['scale'])
    else:
        expected = None
    if expected is not None:
        chi2_stat, chi2_p = stats.chisquare(hist, expected[:-1])
    else:
        chi2_stat, chi2_p = None, None

    return {
        'KS Statistic': ks_stat,
        'KS p-value': ks_p,
        'AD Statistic': ad_stat,
        'AD Critical Values': ad_crit_vals,
        'Chi-Square Statistic': chi2_stat,
        'Chi-Square p-value': chi2_p
    }

# Run GoF tests on the best fit
gof_results = goodness_of_fit_tests(data, best_dist_name, best_params)
print("\nGoodness-of-Fit Results:")
for test, result in gof_results.items():
    print(f"{test}: {result}")

# Plot the best fit
plt.figure(figsize=(10, 6))
sns.histplot(data, kde=False, stat='density', bins=15, label='Data')

# Generate PDF for the best fit
x = np.linspace(min(data), max(data), 100)
if best_dist_name == 'gamma':
    pdf = stats.gamma.pdf(x, a=best_params['a'], loc=best_params['loc'], scale=best_params['scale'])
elif best_dist_name == 'beta':
    pdf = stats.beta.pdf(x, a=best_params['a'], b=best_params['b'], loc=best_params['loc'], scale=best_params['scale'])
elif best_dist_name == 'norm':
    pdf = stats.norm.pdf(x, loc=best_params['loc'], scale=best_params['scale'])

plt.plot(x, pdf, label=f'Best fit: {best_dist_name}', linewidth=2)
plt.title(f'Best Fit: {best_dist_name} (KS p-value = {gof_results["KS p-value"]:.3f})')
plt.xlabel('Attendance Percentage')
plt.ylabel('Density')
plt.legend()
plt.show()
```
